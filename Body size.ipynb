```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rosybarb = body_size[body_size['INDIVIDUAL ID'] == 'RB'] ['BODY SIZE'] 
tigerbarb = body_size[body_size['INDIVIDUAL ID'] == 'TB'] ['BODY SIZE']

fig = plt.figure(figsize=(8, 6))

sns.kdeplot(rosybarb, fill=True, color='red', label = 'Rosy barb')
sns.kdeplot(tigerbarb, fill=True, color='orange', label = 'Tiger barb')

plt.tight_layout()
plt.legend()
plt.show()

fig.savefig('distribution_of_bodysize.png', bbox_inches='tight')
```
