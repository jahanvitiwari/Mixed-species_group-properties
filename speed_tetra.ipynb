```python
ind_speed1 = np.sqrt(vel1[:,:,0]**2 + vel1[:,:,1]**2)
ind_speed2 = np.sqrt(vel2[:,:,0]**2 + vel2[:,:,1]**2)
ind_speed3 = np.sqrt(vel3[:,:,0]**2 + vel3[:,:,1]**2)
ind_speed4 = np.sqrt(vel4[:,:,0]**2 + vel4[:,:,1]**2)
ind_speed5 = np.sqrt(vel5[:,:,0]**2 + vel5[:,:,1]**2)
```

```python
group_speed1 = np.mean(vel1, axis=1)
grp_speed1 = np.sqrt(vel1[:,0]**2 + vel1[:,1]**2)

group_speed2 = np.mean(vel2, axis=1)
grp_speed2 = np.sqrt(vel2[:,0]**2 + vel2[:,1]**2)

group_speed3 = np.mean(vel3, axis=1)
grp_speed3 = np.sqrt(vel3[:,0]**2 + vel3[:,1]**2)

group_speed4 = np.mean(vel4, axis=1)
grp_speed4 = np.sqrt(vel4[:,0]**2 + vel4[:,1]**2)

group_speed5 = np.mean(vel5, axis=1)
grp_speed5 = np.sqrt(vel5[:,0]**2 + vel5[:,1]**2)
```

```python
fig, ax = plt.subplots(1, 5, figsize=(15, 3))

sns.kdeplot(ind_speed1.flatten(), ax=ax[0], fill=True, color='red')
sns.kdeplot(ind_speed2.flatten(), ax=ax[1], fill=True, color='orange')
sns.kdeplot(ind_speed3.flatten(), ax=ax[2], fill=True, color='green')
sns.kdeplot(ind_speed4.flatten(), ax=ax[3], fill=True, color='blue')
sns.kdeplot(ind_speed5.flatten(), ax=ax[4], fill=True, color='indigo')

# Set y-axis limits
for i in range(5):
    ax[i].set_ylim(0, 0.15)
    ax[i].set_xlim(0, 25)

# Adding titles to each subplot
titles = ['ST 16', 'ST 12 + NT 4', 'ST 8 + NT 8', 'ST 4 + NT 12', 'NT 16']
for i in range(5):
    ax[i].set_title(titles[i], fontsize=12, fontweight='bold')

for i in range(5):
    ax[i].set_ylabel('')

# Adding a title for the entire figure
fig.suptitle('Distribution of Individual Speeds (cm/sec)', y=1, fontsize=16, fontweight='bold')

fig.supxlabel(' Individual Speed (cm/sec) ', fontsize=12, fontweight='bold')
fig.supylabel('Density', x=0, fontsize=12, fontweight='bold')

# Adjust layout to prevent overlap
fig.tight_layout()

# Show the plot
fig.show()

fig.savefig('plots_of_individual_speeds_cm.png', bbox_inches='tight')
```
